```python
import pandas as pd
import numpy as np

direc = 'C:/Users/54115/Documents/Ramiro/Taller/'
```

```python
inst = 'tesla'

exp = ['sma','ema', 'boll', 'rsi', 'macd']
```

```python
def read_exp(exp, df):
    if exp == 'sma':
        df['Strategy'] = 'SMA(' + df['short_sma'].astype(str) + ',' + df['long_sma'].astype(str) + ')'
        df = df[['Strategy', 'pnl', 'port_max_dd']]
        
    if exp == 'ema':
        df['Strategy'] = 'EMA(' + df['short_ema'].astype(str) + ',' + df['long_ema'].astype(str) + ')'
        df = df[['Strategy', 'pnl', 'port_max_dd']]
        
    if exp == 'boll':
        df['Strategy'] = 'BOLL(' + df['ma'].astype(str) + ',' + df['stdev'].astype(str) + ')'
        df = df[['Strategy', 'pnl', 'port_max_dd']]
        
    if exp == 'rsi':
        df['Strategy'] = 'RSI(' + df['periods'].astype(str) + ',' + df['oversold'].astype(str) + ',' + df['overbought'].astype(str) +')'
        df = df[['Strategy', 'pnl', 'port_max_dd']]
        
    if exp == 'macd':
        df['Strategy'] = 'MACD(' + df['short_ema'].astype(str) + ',' + df['long_ema'].astype(str) + ',' + df['signal'].astype(str) +')'
        df = df[['Strategy', 'pnl', 'port_max_dd']]
        
    return df
```

```python
def transform_exp(inst):
    result_df = pd.DataFrame()
    
    for i in range (0,5):
        df = pd.read_excel(direc + 'experiment_' + exp[i] + '_' + inst + '.xlsx')
        df = read_exp(exp[i], df)
        
        result_df = result_df.append(df)
        
    result_df.reset_index(inplace = True, drop = True)
    result_df.sort_values(by='pnl', ascending = False, inplace = True)
    
    name_col = 'pnl_' + inst
    result_df.rename(columns={"pnl": name_col}, inplace = True)
    
    result_df.to_excel('C:/Users/54115/Documents/Ramiro/Taller/result' + inst + '.xlsx', index = False)
        
    return result_df
```

```python
transform_exp(inst)
```

```python
final_result = transform_exp('all').merge(transform_exp('microstrategy'), left_on = 'Strategy', right_on= 'Strategy')
final_result = final_result.merge(transform_exp('tesla'), left_on = 'Strategy', right_on= 'Strategy')
final_result = final_result.merge(transform_exp('salvador'), left_on = 'Strategy', right_on= 'Strategy')

final_result.sort_values(by=['pnl_all', 'pnl_microstrategy', 'pnl_tesla', 'pnl_salvador'], ascending = False, inplace = True)

final_result = final_result[['Strategy', 'pnl_all', 'pnl_microstrategy', 'pnl_tesla', 'pnl_salvador']]

final_result.to_excel('C:/Users/54115/Documents/Ramiro/Taller/final_result.xlsx', index = False)

final_result
```
